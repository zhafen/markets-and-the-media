# Aplying threshhold filtering to the data

```python
## Loading test and train media data
import pandas as pd

media_test = pd.read_csv("../data/test/media.csv", index_col = 0, parse_dates = ['pub_date'])
media_train = pd.read_csv("../data/train/media.csv", index_col = 0, parse_dates = ['pub_date'])
```

```python
media_test.describe()
```

```python
media_test.sample(5)
```

```python
media_train.describe()
```

```python
## use nltk for word count

from textblob import TextBlob
```

```python
## set word count threshold

word_threshold = 1
word = 'google'
```

```python
## drop rows in  media_test where the 'text' column mentions the variable word less than specified variable the word_threshold

media_test = media_test[media_test['text'].apply(lambda x: TextBlob(x).words.count(word) >= word_threshold)]

## drop rows in  media_train where the text column has word less than the word_threshold

media_train = media_train[media_train['text'].apply(lambda x: TextBlob(x).words.count(word) >= word_threshold)]

```

```python
media_test.sample(5)
```

```python
media_train.describe()
```

```python
### save the filtered data

media_test.to_csv("../data/test/media_filtered.csv")
media_train.to_csv("../data/train/media_filtered.csv")
```
